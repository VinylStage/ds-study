# Q2
2011년부터의 연도별, 국가별 평균 기대수명

```python
# groupby(unstack()) or pivot_table 사용하기
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

```python
# DataFrame 정보 확인
df.info()
```

```python
# 'Spending_USD'행 삭제
df1 = df.drop(columns=['Spending_USD'])
df1.head()
```

```python
# 'Year'이 2011 이후부터만 남게
df2 = df1[df1['Year'] >= 2011]
df2.head()
```

## 1) Pivot table


```python
pd.pivot_table(df2, index='Year', columns='Country', values='Life_Expectancy')
```

## 2) Groupby

```python
df2.groupby(["Year", "Country"])["Life_Expectancy"].mean().unstack()
```

### 2.3 파일 경로 설정 방법

### 1 절대 경로

* 절대경로는 어떠한 웹페이지나 파일이 가진 고유한 경로로 최상위 디렉토리가 포함된 경로
* 예) file:///Users/user/Downloads/
* 예) C:\user\documents

### 2 상대 경로

* 상대경로는 현재 위치한 곳을 기준으로 하는 경로입니다.
* /는 루트
* ./는 현재 위치
* ../는 현재 위치의 상위 폴더
* ../../는 두 단계 위의 상위 폴더를 의미합니다.

#### 파일 하나 불러오기
* data/ + tab 눌렀을 때 폴더 내 데이터가 하나면 해당 파일을 불러옴
* UnicodeDecodeError - 파일은 잘 불러왔는데 인코딩 에러가 발생함 -> 인코딩 설정을 해야함
* FileNotFoundError - 파일 이름이 잘못 들어감
* cp949는 한글 윈도우에서 사용되는 인코딩 방식.
* 기본적으로 판다스는 utf 파일을 사용하기 때문에 잘 불러올 수 있음
* data 폴더를 적지 않고, 파일이름만 적었을 때 주피터 노트북이 있는 경로에 있으면 불러지지만 아닐 경우 (ex. 폴더 내)에는 불러와지지 않는다.
* glob : 유닉스 스타일의 규칙을 통해서 경로명을 찾을 수 있다. 특정 경로에 어떤 파일이 있는지 확인 가능.

### 3 여러 파일을 한번에 로드하는 방법
